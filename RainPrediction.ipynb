```python
# Slicing Essential Data
features_considered = ['cl', 'albedo_5', 'albedo_6', 'tbb_7', 'tbb_11', 'tbb_12', 'tbb_13']
features = data[features_considered]
features.index = data[['date','time']]
features.plot(subplots=True)
```

```python
# Find all columns with NaN values
nan_cols = features.columns[features.isna().any()].tolist()

# Fill NaN values with the mean of the corresponding column
for col in nan_cols:
    features[col].fillna(features[col].mean(), inplace=True)

# another sanity check to make sure that there are not more any nan
print(features.isnull().sum())
```

#Processing

```python
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
X = features.values
y = data.label
```

```python
# Processing 1 - Features Preparation

rows = []
for i in range(len(X)-5):
    row = []
    for j in range(i, i+6):
        row.append(X[j])
    rows.append(row)


X = np.array(rows)

print(X.shape) # contohnya: 3, 6, 7 -> 3 rows data, 6 timeframe, 7 features. Nah harusnya 6 timeframe sama 7 features bakal tetap, cuma rowsnya nanti nambah karnea di looping sampai akhir data
```

```python
# Processing 2 - Label Preparation

''' 
Nah karena datanya digabung per 6 timeframe, labelnya juga harus digabung per 6 timeframe
Kyknya better kalau jgn dipisah dlu X sama y nya yang kyk di code ini.

Better kalau dari dataset keseluruhan, diwindow dulu baru dipisah nanti X sama y nya

Kalau untuk per row kan tinggal append ke array aja ya (kyk di definisikan sblmnya)

Untuk yang label bakal di merge jadi 1. Caranya tinggal di average aja. Kalau hasil avgnya <0.5 maka dianggap 0 (tidak hujan)
kalau >=0.5 dianggap 1 (hujan)
'''

# map the label value into 0 & zero
mapping = {"hujan deras": 1, "tidak hujan": 0}
updated_label = np.array([mapping[val] for val in y])

# group the label by 6 timeframes
rows = []
for i in range(len(updated_label)-5):
    row = []
    for j in range(i, i+6):
        row.append(updated_label[j])
    rows.append(row)

grouped_label = np.array(rows)

# create mapping dictionary for map the mean of each 6 data
mapping = {True: 1, False: 0}
y = np.array([mapping[val >= 0.5] for val in np.mean(grouped_label, axis=1)])

unique, counts = np.unique(y, return_counts=True)

for value, count in zip(unique, counts):
    print(f"{value}: {count}")
print(y.shape)
```

```python
# Processing 2 - Label Preparation (2)

''' 
Label yang sudah ada di onehotencoding:
* 0 menjadi [1,0]
* 1 menjadi [0,1]
'''

num_classes = 2
y = np.eye(num_classes)[y]

print(y.shape)
print(y[0])
```

```python
# Processing 3 - Train Test Split

''' 
Pisahkan X dan y nya

X itu untuk data yang berisi fiturnya

y itu untuk data labelnya

Di train test split jadi X_train, y_train, X_test, y_test
% train = 80%
% test = 20%
'''
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=27)

# show splitted data shape
print("Shape X_train: ", X_train.shape)
print("Shape y_train: ", y_train.shape)
print("Shape X_test: ", X_test.shape)
print("Shape y_test: ", y_test.shape)
```

#Model Training

## LSTM

```python
from keras.models import Sequential
from keras.layers import LSTM, Dense
```

```python
# Define the model architecture
modelLSTM = Sequential()
modelLSTM.add(LSTM(50, input_shape=(X_train.shape[1], X_train.shape[2])))
modelLSTM.add(Dense(2, activation='softmax'))

# Compile the model
modelLSTM.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model
historyLSTM = modelLSTM.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test))
```

```python
# Plot the train and validation loss on the left side
plt.subplot(1, 2, 1)
plt.plot(historyLSTM.history['loss'], label='train_loss')
plt.plot(historyLSTM.history['val_loss'], label='val_loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

# Plot the train and validation accuracy on the right side
plt.subplot(1, 2, 2)
plt.plot(historyLSTM.history['accuracy'], label='train_acc')
plt.plot(historyLSTM.history['val_accuracy'], label='val_acc')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
```

##RNN

```python
from keras.models import Sequential
from keras.layers import SimpleRNN, Dense
```

```python
# Define the model architecture
modelRNN = Sequential()
modelRNN.add(SimpleRNN(50, input_shape=(X_train.shape[1], X_train.shape[2])))
modelRNN.add(Dense(2, activation='softmax'))

# Compile the model
modelRNN.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model
historyRNN = modelRNN.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test))
```

```python
# Plot the train and validation loss on the left side
plt.subplot(1, 2, 1)
plt.plot(historyRNN.history['loss'], label='train_loss')
plt.plot(historyRNN.history['val_loss'], label='val_loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

# Plot the train and validation accuracy on the right side
plt.subplot(1, 2, 2)
plt.plot(historyRNN.history['accuracy'], label='train_acc')
plt.plot(historyRNN.history['val_accuracy'], label='val_acc')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
```
